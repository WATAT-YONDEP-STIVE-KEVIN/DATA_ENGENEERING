<center><h3>affichons le contenu de la page web avec BeautifulSoup de python</h3></center>

```python
from bs4  import BeautifulSoup

#Strucutrer les pages web selon la structure html
pageWeb1Soup = BeautifulSoup(pageWeb1.content, 'html.parser')
webPage2Soup = BeautifulSoup(webPage2.content, "html.parser")
```

```python
print(pageWeb1Soup.prettify())#afficher le contenu de la page web
```

```python
print(webPage2Soup.prettify())
```

```python
"""
Si nous voulons afficher le contenu d'un element de la page on utilisera soup.find_all('Nom_element_html')
"""

valeur = pageWeb1Soup.find_all("t#mw-content-text > div.mw-content-ltr.mw-parser-output > table.wikitable.sortable.sticky-header-multi.static-row-numbers.jquery-tablesorter > tbody > tr:nth-child(27) > td:nth-child(5)")
print(valeur)


#nous pouvons voir q'aucun element ne s'affiche alors que sur le site web alors qu'on a cibler
# une valeur precise dans la page web
#Le fait que rien ne s'affiche est du au fait que wikipedia limite le web scraping comme je l'ao dit dans le support
#Donc dans la suite nous nous contenterons d'utiliser webPage2soup pour effectuer notre travail
```

```python
print(webPage2Soup.prettify())
```

```python
jourSept = webPage2Soup.find(id="seven-day-forecast") #recuperons les previsions metéo le 7eme jour
print(jourSept)
```

```python
"""
le rendu du code precedant nous donne des elements html
ainsi pour avoir les previsions du 7eme jour on va recuperer l'element html
ayant la class="tombstone-container" qui sont dans jourSept
"""
```

```python
previsions_items = jourSept.find_all(class_="tombstone-container")
print(previsions_items)
#la methode find_all retourne un tableau d'element dont on peut avoir acces par indicage
```

```python
previsions_nuit = previsions_items[0]
print(previsions_nuit.prettify())
```

Maintenant faisons l'extration des informations sur la prevision de la soirée le jour 7

```python
periode = previsions_nuit.find(class_="period-name").get_text()#pour recuperer le contenu de l'element ciblé
etat_periode = previsions_nuit.find(class_="short-desc").get_text()
temperature = previsions_nuit.find(class_="temp").get_text()

print(f"periode : {periode}")
print("etat de la periode : {}".format(etat_periode))
print("temperature : ", temperature)
```

selectionnons maintenant tous les elements ou periodes ciblés par "tombstone-container" le septieme jour , pour cela on peut utiliser facilement les liste en comprehension de python

```python
periode_tag = jourSept.select(".tombstone-container .period-name")
print(periode_tag)
periode = [pt.get_text() for pt in periode_tag]
print(periode)
```

```python
[print(periode[i]) for i in range(len(periode)) ]
```

on peut appliquer la meme technique precedente pour obtenir les autres données

```python
etats_periode = [etat.get_text() for etat in jourSept.select(".tombstone-container .short-desc")]
temperatures = [temp.get_text() for temp in jourSept.select(".tombstone-container .temp")]
descs = [desc["title"] for desc in jourSept.select('.tombstone-container img')]

print(etats_periode)
print(temperature)
print(descs)
```

Maintenant stockons nos données scrapées dans un fichier csv

```python
import pandas as pd

meteo = pd.DataFrame({"periode":etats_periode,
"short_desc": periode,
"temperaure":temperature,
"desc": descs})
```

```python
meteo.to_csv('./datas/dataScraping.csv')
```
